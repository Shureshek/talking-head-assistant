```python
from qwen_tts import Qwen3TTSModel
import soundfile as sf
```

```python


model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-0.6B-Base",
    device_map="cuda:0",
    dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
)

wavs, sr = model.generate_voice_clone(
    text="Я бля сегодня бля весь день бля работал и конкретно бля заебался. Всё, что я сейчас бля хочу - это набухаться бля и съесть бля огромного вкусного мамонта бля. Вот.",
    ref_audio="audio/Vasya/audio_2026-01-30_17-49-34.ogg",
    ref_text="Помню, что я два... короче, перед того, как к тебе приехать, я со сутки или... не помню, короче, в кафе покупал кофе и что-то доставал, их щупал, ну то есть, я их обратно положил, наверно, два дня назад или три, не помню, потом я приехал... ну, то есть, получается, я с тачки не выходил, мы катались с девочкой, потом переехал домой, днём поехал к тебе, либо они... блядь, либо... хуй, я не знаю, где они, короче, либо в тачке, но тачки сейчас у меня нет, она в сервисе, блядь, либо нужно туда идти, блядь.",   # помогает
    language="Russian"
)
sf.write("output_voice_clone.wav", wavs[0], sr)
```
